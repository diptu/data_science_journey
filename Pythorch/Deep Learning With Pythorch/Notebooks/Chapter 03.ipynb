```python
l = list(range(6))
print(l[:])           # All elements in the list
print(l[1:4])         # From element 1 inclusive to elemnt 4 exclusive
print(l[1: ])         # From element 1 inclusive to the end to the list  
print(l[ :4])         # From the start of the list to element 4 exclusive
print(l[ :-1])        # From the start of the list to one before last element
print(l[ 1: 4 : 2])   # From element 1 to 4 exclusive in step of 2 
```

```python
img_t = torch.randn(3,5,5)                        #shape [chanel, rows, columns]
weights = torch.tensor([0.2126, 0.7152, 0.0722])
batch_t = torch.randn(2, 3, 5,5)                  #shape [batch, chanel, rows, columns]
print(img_t)
print(weights)
print(batch_t)
```

```python
img_gray_naive = img_t.mean(-3)                   # Chanel always in dimention -3
batch_gray_naive = batch_t.mean(-3)

print(img_gray_naive)
print(batch_gray_naive)

print(img_gray_naive.shape, batch_gray_naive.shape)
```

```python
unsqueezed_weights = weights.unsqueeze(-1).unsqueeze_(-1)
print(unsqueezed_weights)
print(unsqueezed_weights.shape)
```

```python
img_weights = img_t * unsqueezed_weights
batch_weights = batch_t * unsqueezed_weights
img_gray_weighted = img_weights.sum(-3)
batch_gray_weighted = batch_weights.sum(-3)

print(img_weights)
print(batch_weights)
print(img_gray_weighted)
print(batch_gray_weighted)

print(batch_weights.shape, batch_t.shape, unsqueezed_weights.shape)
```

```python
img_named = img_t.refine_names(..., 'chanels', 'rows','columns')
batch_named = batch_t.refine_names(..., 'chanels', 'rows','columns')

print(img_named)
print(batch_named)

print(img_named.shape, img_named.names)
print(batch_named.shape, batch_named.names)
```
